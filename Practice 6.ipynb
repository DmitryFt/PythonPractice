```python
import json
with open('comments_326022.json', 'r') as f:
     text = f.read()
info = json.loads(text) #начальный словарь с ключами "note" и "comments"
info.keys() #его ключи
# но значение ключа comments - список* словариков!
spisok = info['comments'] # значение этого ключа - получили список
# теперь надо рассмотреть эти мелкие словарики
# т.е. каждый словарик - есть элемент списка
# каждый элемент можем вынести через цикл
# каждый элемент списка "comments" - словарик с двумя парами "ключ(name) - значение(count)"
# Соберите словарь, ключами в котором будут имена пользователей - (name), 
# а значениями - количество сделанных комментариев(count)
# в итоге будет: словарь с элементами "name - count"
Res = {}
total = 0
for elem in spisok:
    #print(elem.items())
#     items = [(elem.get('name'), elem.get('count'))]
#     Res = dict(items)
    #Res.update {elem.get('name') : elem.get('count')}
    Res[elem['name']] = elem.get('count')
    count = int(elem.get('count'))
    total += count
print("total: ", total)
Res
```

### Задание 2:
Выполните то же задание, что и в предыдущем пункте, только теперь исходные данные содержатся в формате xml. 

```python
import xml.etree.ElementTree as ET
Res = {}
total = 0
tree = ET.parse('comments_326018.xml')
root = tree.getroot()
for children in root:
    for elem in children.findall("comment"):
        count = int(elem.find('count').text)
        Res[elem.find('name').text] = count
        total += count

print("total: ", total)
Res
```
